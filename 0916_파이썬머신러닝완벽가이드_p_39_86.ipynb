### DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환

1. 넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기

   !! 2차원 이하의 데이터들만 DataFrame으로 변환될 수 있음 !!

> 1차원 형태의 리스트와 넘파이 ndarray -> DataFrame 변환

````python
import numpy as np

col_name1=['col1']
list1=[1,2,3]
array1=np.array(list1)
print('array1 shape:',array1.shape)

# 리스트를 이용해 DataFrame 생성
df_list1=pd.DataFrame(list1,columns=col_name1)
print('1차원 리스트로 만든 DataFrame:',df_list1)

# 넘파이 ndarray를 이용해 DataFrame 생성
df_array1=pd.DataFrame(array1,columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:',df_array1)
````

> 2차원 형태의 데이터 -> DataFrame 생성

````python
# 3개의 column명이 필요함
col_name2=['col1','col2','col3']

# 2행x3열의 리스트와 ndarray를 생성한 뒤 이를 DataFrame으로 변환
list2=[[1,2,3],
       [11,12,13]]
array2=np.array(list2)
print('array2 shape:',array2.shape)

df_list2=pd.DataFrame(list2,columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n',df_list2,'\n')

df_array2=pd.DataFrame(array2,columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n',df_array2)
````

 > 딕셔너리를 DataFrame으로 변환

 - 딕셔너리의 키(key)는 칼럼명으로
 - 딕셔너리의 값(value)는 키에 해당하는 칼럼 데이터로 변환

 =>  키의 경우는 문자열, 값의 경우 리스트(또는 ndarray) 형태로 딕셔너리를 구성
 

````python
# Key는 문자열 column명으로 매핑, Value는 리스트 형(또는 ndarray) column 데이터로 매핑
dict={'col1':[1,11],'col2':[2,22],'col3':[3,33]}
print(dict,'\n')

df_dict=pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n',df_dict)
````

DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환하기

> DataFrame을 넘파이 ndarray로 변환
- DataFrame 객체의 values를 이용

````python
# DataFrame(df_dict)을 ndarray로 변환
array3=df_dict.values
print('df_dict,values 타입:',type(array3),'df_dict.values shape:',array3.shape)
print(array3)
````

> DataFrame을 리스트와 딕셔너리로 변환
* 리스트로의 변환 : values로 얻은 ndarray에 tolist( )를 호출
* 딕셔너리로의 변환 : DataFrame 객체의 to_dict( )메서드 호출, 인자로 'list' 입력 -> 딕셔너리의 값이 리스트혀

````python
# DataFrame을 list로 변환
list3=df_dict.values.tolist()
print('df_dict.values.tolist() 타입:',type(list3))
print(list3)
````

````python
# DataFrame을 딕셔너리로 변환
dict3=df_dict.to_dict('list') # 인자로 'list'를 입력 시, 딕셔너리의 값(value)이 리스트형으로 반환됨
print('df_dict.to_dict() 타입:',type(dict3))
print(dict3)
````

> loc[ ] 에 슬라이싱 기호 : 적용시 유의 사항
- 일반적인 슬라이싱 : 시작값 ~ 종료값 - 1 (종료값 바로 직전까지)
- loc[ ]의 슬라이싱 : 시작값 ~ 종료값 (종료값 까지 포함)
  * 명칭은 숫자형이 아닐 수도 있기 때문에 -1을 할 수 없음

````python
print('위치 기반 iloc slicing\n',data_df.iloc[0:1,0],'\n')
print('명칭 기반 loc slicing\n',data_df.loc['one':'two','Name'])
````

정리 ) iloc 와 loc 의 문제점과 주의사항
- iloc[]는 위치 기반 인덱싱만 가능 
  - 행과 열 위치 값으로 정수형 값을 지정해 원하는 데이터 반환
- loc[]는 명칭 기반 인덱싱만 가능
  - 행 위치에 데이터프레임의 인덱스가 옴, 열 위치에는 칼럼 명을 지정해 원하는 데이터 반환
*  명칭 기반 인덱싱에서 슬라이싱을 '시작점:종료점'으로 지정시, 시작점에서 종료점을 포함한 위치에 있는 데이터를 반환

### apply lamda 식으로 데이터 가공
: apply 함수에 lamda 식을 결합해 데이터프레임이나 시리즈의 레코드별로 데이터를 가공하는 기능 제공

````python
def get_square(a):
  return a**2
print('3의 제곱은:',get_square(3))
````

> lambda : 함수의 선언과 함수 내의 처리를 한줄의 식으로 쉽게 변환

````python
lambda_square=lambda x : x**2 # ':'로 입력인자와 반환될 입력인자의 계산식 분리
print('3의 제곱은:',lambda_square(3))
````

> 여러 개의 값을 입력 인자로 사용해야 할 경우, map() 함수 결합

````python
a=[1,2,3]
squares=map(lambda x: x**2, a)
list(squares)
````

> 데이터프레임의 apply에 lamda 식 적용한 데이터 가공

````python
# 'Name' 칼럼의 문자열 개수를 별도의 칼럼인 'Name_len'에 생성
titanic_df['Name_len']=titanic_df['Name'].apply(lambda x: len(x))
titanic_df[['Name','Name_len']].head(3)
````

> lamda 식에서 if else 절 사용한 복잡한 가공

*   if절 : if 식보다 반환 값ㅇ을 먼저 기술해야 함 ( lamda 식 ':' 기호의 오른편에 반환 값이 있어야 함)
* if, else 만 지원하고 else if 는 지원하지 않음
  - else 절을 내포해 () 내에서 다시 if else 적용해야 함

````python
# 나이가 15세 미만이면 'Child', 그렇지 않으면 'Adult'로 구분
titanic_df['Child_Adult']=titanic_df['Age'].apply(lambda x: 'Child' if x <=15 else 'Adult')
titanic_df[['Age','Child_Adult']].head(8)
````

````python
# 나이가 15세 이하면 Child, 15세~60세 사이는 Adult, 61세 이상은 Elderly로 분류하는 'Age_Cat' 칼럼 만들기
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x: 'Child' if x <=15 else ('Adult' if x <=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
````

````python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age):
  cat=''
  if age <= 5: cat='Baby'
  elif age <= 12: cat='Child'
  elif age <= 18: cat='Teenager'
  elif age <= 25: cat='Student'
  elif age <= 35: cat='Young Adult'
  elif age <= 60: cat='Adult'
  else: cat='Elderly'
  return cat

# lambda 식에서 위에서 생성한 get_category() 함수를 반환값으로 지정
````

````python
# get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age','Age_cat']].head()
````
